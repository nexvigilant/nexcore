## 2. The Five Axioms

Five axioms govern all state machines:

| Axiom | Name | Statement |
|:-----:|------|----------|
| **A1** | Finite Decomposition | Every state machine has finitely many states |
| **A2** | Hierarchical Order | States admit a partial order (DAG or cycles) |
| **A3** | Conservation | Transitions preserve designated invariants |
| **A4** | Safety Manifold | Valid states form interior; terminal states form boundary |
| **A5** | Emergence | Complex behavior emerges from simple transition rules |

````python
from abc import ABC, abstractmethod
from typing import Generic, TypeVar, List, Set

S = TypeVar('S')  # State type

class Axiom(ABC):
    """Base class for axiom witnesses."""
    
    @staticmethod
    @abstractmethod
    def id() -> str: ...
    
    @staticmethod
    @abstractmethod
    def name() -> str: ...
    
    @staticmethod
    @abstractmethod
    def statement() -> str: ...
    
    @abstractmethod
    def witness(self) -> str: ...


@dataclass
class A1_FiniteDecomposition(Axiom):
    """Axiom A1: Every state machine has finitely many states."""
    max_states: int
    actual_count: int
    
    @staticmethod
    def id() -> str: return "A1"
    
    @staticmethod
    def name() -> str: return "Finite Decomposition"
    
    @staticmethod
    def statement() -> str:
        return "Every state machine has finitely many states (bounded by MAX)"
    
    def witness(self) -> str:
        return f"|S| = {self.actual_count} ≤ {self.max_states} = MAX"
    
    def is_satisfied(self) -> bool:
        return 0 < self.actual_count <= self.max_states


@dataclass
class A4_SafetyManifold(Axiom):
    """Axiom A4: Valid states form interior; terminal states form boundary."""
    interior_states: Set[str]   # Non-terminal states
    boundary_states: Set[str]   # Terminal states (absorbing)
    exterior_states: Set[str]   # Invalid/error states
    
    @staticmethod
    def id() -> str: return "A4"
    
    @staticmethod
    def name() -> str: return "Safety Manifold"
    
    @staticmethod
    def statement() -> str:
        return "Valid states form interior; terminal states form absorbing boundary"
    
    def witness(self) -> str:
        return f"Interior: {self.interior_states}, Boundary: {self.boundary_states}"
    
    def is_partitioned(self) -> bool:
        """Check that interior and boundary are disjoint."""
        return len(self.interior_states & self.boundary_states) == 0


# Example: Traffic Light State Machine
print("Example: Traffic Light Axiom Verification")
print("=" * 45)

a1 = A1_FiniteDecomposition(max_states=10, actual_count=3)
print(f"\n{a1.id()}: {a1.name()}")
print(f"  Statement: {a1.statement()}")
print(f"  Witness: {a1.witness()}")
print(f"  Satisfied: {a1.is_satisfied()}")

a4 = A4_SafetyManifold(
    interior_states={'red', 'yellow', 'green'},
    boundary_states={'off'},
    exterior_states={'broken'}
)
print(f"\n{a4.id()}: {a4.name()}")
print(f"  Statement: {a4.statement()}")
print(f"  Witness: {a4.witness()}")
print(f"  Partitioned: {a4.is_partitioned()}")
````

## 3. Conservation Laws

Three fundamental conservation laws that every state machine must satisfy:

| Law | Name | Statement |
|:---:|------|----------|
| **L3** | Single State | Machine is in exactly one state at any time |
| **L4** | Non-Terminal Flux | Non-terminal states have ≥1 outgoing transition |
| **L11** | Structure Immutability | State count is immutable after construction |

````python
@dataclass
class LawVerification:
    """Result of verifying a conservation law."""
    law_id: str
    satisfied: bool
    witness: str
    counterexample: Optional[str] = None


class L3_SingleState:
    """Law L3: Machine is in exactly one state at any time."""
    
    @staticmethod
    def verify(active_state_count: int) -> LawVerification:
        satisfied = active_state_count == 1
        return LawVerification(
            law_id="L3",
            satisfied=satisfied,
            witness=f"active_states = {active_state_count}",
            counterexample=None if satisfied else f"Expected 1, got {active_state_count}"
        )


class L4_NonTerminalFlux:
    """Law L4: Non-terminal states have at least one outgoing transition."""
    
    @staticmethod
    def verify(is_terminal: bool, outgoing_count: int) -> LawVerification:
        # Terminal states can have 0 outgoing; non-terminal must have ≥1
        satisfied = is_terminal or outgoing_count >= 1
        return LawVerification(
            law_id="L4",
            satisfied=satisfied,
            witness=f"terminal={is_terminal}, outgoing={outgoing_count}",
            counterexample=None if satisfied else "Non-terminal state with no outgoing transitions"
        )


class L11_StructureImmutability:
    """Law L11: State count is immutable after construction."""
    
    @staticmethod
    def verify(initial_count: int, current_count: int) -> LawVerification:
        satisfied = initial_count == current_count
        return LawVerification(
            law_id="L11",
            satisfied=satisfied,
            witness=f"initial={initial_count}, current={current_count}",
            counterexample=None if satisfied else f"State count changed from {initial_count} to {current_count}"
        )


# Verify conservation laws
print("Conservation Law Verification")
print("=" * 45)

laws = [
    L3_SingleState.verify(active_state_count=1),
    L4_NonTerminalFlux.verify(is_terminal=False, outgoing_count=2),
    L11_StructureImmutability.verify(initial_count=5, current_count=5),
]

for v in laws:
    status = "✓" if v.satisfied else "✗"
    print(f"  {status} {v.law_id}: {v.witness}")
````

## 4. Typestate Pattern

The **typestate pattern** enforces state machine correctness at compile time:

- States are represented as **zero-sized types** (ZSTs)
- Transitions are **methods** that consume `self` and return a new type
- Invalid transitions cause **compile errors**, not runtime errors

```rust
// Rust typestate example
struct Switch<S: State> {
    _state: PhantomData<S>,
}

impl Switch<Off> {
    fn turn_on(self) -> Switch<On> { ... }
}

impl Switch<On> {
    fn turn_off(self) -> Switch<Off> { ... }
}

// Switch<Off>.turn_off() → COMPILE ERROR (no such method)
```

````python
from typing import TypeVar, Generic, Callable

# Python approximation of typestate using generics
S = TypeVar('S')  # State type
T = TypeVar('T')  # Payload type

class TypesafeWrapper(Generic[S, T]):
    """A wrapper that tracks state at the type level."""
    
    def __init__(self, state_name: str, payload: T):
        self._state_name = state_name
        self._payload = payload
        self._transition_count = 0
    
    @property
    def state_name(self) -> str:
        return self._state_name
    
    @property
    def payload(self) -> T:
        return self._payload
    
    @property
    def transition_count(self) -> int:
        return self._transition_count


class TransitionBuilder(Generic[S, T]):
    """Builder for fluent transitions with payload transformation."""
    
    def __init__(self, wrapper: TypesafeWrapper[S, T]):
        self._wrapper = wrapper
        self._transform: Callable[[T], T] = lambda x: x
    
    def map(self, f: Callable[[T], T]) -> 'TransitionBuilder[S, T]':
        """Apply transformation to payload during transition."""
        self._transform = f
        return self
    
    def to(self, new_state: str) -> TypesafeWrapper:
        """Complete the transition to a new state."""
        new_wrapper = TypesafeWrapper(
            state_name=new_state,
            payload=self._transform(self._wrapper.payload)
        )
        new_wrapper._transition_count = self._wrapper._transition_count + 1
        return new_wrapper


# Example: Order lifecycle
print("Typestate Pattern Example: Order Lifecycle")
print("=" * 50)

# Create order in 'Pending' state
order = TypesafeWrapper(state_name="Pending", payload={"items": ["book", "pen"], "total": 25.00})
print(f"Initial: {order.state_name} → {order.payload}")

# Transition: Pending → Confirmed (apply discount)
order = TransitionBuilder(order).map(lambda p: {**p, "total": p["total"] * 0.9}).to("Confirmed")
print(f"After confirm: {order.state_name} → {order.payload}")

# Transition: Confirmed → Shipped
order = TransitionBuilder(order).map(lambda p: {**p, "tracking": "TRK123"}).to("Shipped")
print(f"After ship: {order.state_name} → {order.payload}")

# Transition: Shipped → Delivered (terminal)
order = TransitionBuilder(order).to("Delivered")
print(f"Final: {order.state_name} (transitions: {order.transition_count})")
````

## 5. Algebraic Operations

State machines form an algebra with four fundamental operations:

| Operation | Notation | Meaning |
|-----------|:--------:|----------|
| **Product** | M₁ × M₂ | Run M₁ and M₂ in parallel |
| **Coproduct** | M₁ + M₂ | Choose between M₁ or M₂ |
| **Sequence** | M₁ ; M₂ | Run M₁ then M₂ |
| **Iteration** | M* | Run M zero or more times |

````python
from dataclasses import dataclass
from typing import Tuple, Union, List

@dataclass
class SimpleState:
    name: str
    is_terminal: bool = False
    is_initial: bool = False


class StateProduct:
    """
    Product of two state machines (parallel composition).
    
    States are pairs (s₁, s₂).
    Both machines run simultaneously.
    """
    
    def __init__(self, left: List[SimpleState], right: List[SimpleState]):
        self.left = left
        self.right = right
    
    def state_count(self) -> int:
        """Product has |L| × |R| states."""
        return len(self.left) * len(self.right)
    
    def states(self) -> List[Tuple[SimpleState, SimpleState]]:
        """Generate all product states."""
        return [(l, r) for l in self.left for r in self.right]
    
    def is_terminal(self, pair: Tuple[SimpleState, SimpleState]) -> bool:
        """Product state is terminal iff BOTH components are terminal."""
        return pair[0].is_terminal and pair[1].is_terminal


class StateCoproduct:
    """
    Coproduct of two state machines (choice composition).
    
    States are tagged union: Left(s₁) | Right(s₂).
    Choose one machine to run.
    """
    
    def __init__(self, left: List[SimpleState], right: List[SimpleState]):
        self.left = left
        self.right = right
    
    def state_count(self) -> int:
        """Coproduct has |L| + |R| states."""
        return len(self.left) + len(self.right)
    
    def is_terminal(self, state: Union[Tuple[str, SimpleState], None]) -> bool:
        """Coproduct state is terminal iff the component is terminal."""
        if state is None:
            return False
        tag, s = state
        return s.is_terminal


# Example: Product of two simple machines
print("Algebraic Operations Example")
print("=" * 50)

# Machine A: {A0, A1}
machine_a = [SimpleState("A0", is_initial=True), SimpleState("A1", is_terminal=True)]

# Machine B: {B0, B1, B2}
machine_b = [SimpleState("B0", is_initial=True), SimpleState("B1"), SimpleState("B2", is_terminal=True)]

# Product: A × B
product = StateProduct(machine_a, machine_b)
print(f"\nProduct A × B:")
print(f"  States: |A| × |B| = {len(machine_a)} × {len(machine_b)} = {product.state_count()}")
print(f"  All states: {[(s[0].name, s[1].name) for s in product.states()]}")

# Coproduct: A + B
coproduct = StateCoproduct(machine_a, machine_b)
print(f"\nCoproduct A + B:")
print(f"  States: |A| + |B| = {len(machine_a)} + {len(machine_b)} = {coproduct.state_count()}")
````

## 6. Category Theory Foundations

State machines form a category **StateMach** where:

- **Objects**: State machine types (finite sets of states + transitions)
- **Morphisms**: Structure-preserving maps (simulations)
- **Identity**: Identity simulation
- **Composition**: Sequential simulation composition

### Key Functors

| Functor | Domain | Codomain | Meaning |
|---------|--------|----------|---------|
| Trace | StateMach | Set | Extracts trace language |
| Reach | StateMach | Set | Extracts reachable states |
| Obs | StateMach | StateMach | Observable behavior |

````python
@dataclass
class StateMachineSpec:
    """A state machine specification (object in StateMach category)."""
    id: int
    states: List[str]
    initial: int
    terminals: List[int]
    transitions: int = 0
    
    @property
    def state_count(self) -> int:
        return len(self.states)


@dataclass
class Simulation:
    """
    A simulation between state machines (morphism in StateMach).
    
    A simulation R ⊆ S₁ × S₂ is a relation such that:
    1. (s₁_init, s₂_init) ∈ R (initial states related)
    2. If (s₁, s₂) ∈ R and s₁ →ᵃ s₁', then ∃s₂' such that s₂ →ᵃ s₂' and (s₁', s₂') ∈ R
    """
    source_id: int
    target_id: int
    state_map: List[int]  # source state → target state
    is_bisimulation: bool = False
    
    @classmethod
    def identity(cls, machine_id: int, state_count: int) -> 'Simulation':
        """Identity simulation (category identity morphism)."""
        return cls(
            source_id=machine_id,
            target_id=machine_id,
            state_map=list(range(state_count)),
            is_bisimulation=True
        )
    
    def compose(self, other: 'Simulation') -> Optional['Simulation']:
        """Compose simulations: (f ; g) where f: A → B and g: B → C."""
        if self.target_id != other.source_id:
            return None
        
        composed_map = [other.state_map[s] for s in self.state_map]
        return Simulation(
            source_id=self.source_id,
            target_id=other.target_id,
            state_map=composed_map,
            is_bisimulation=self.is_bisimulation and other.is_bisimulation
        )


# Example: Simulation composition
print("Category Theory: Simulation Composition")
print("=" * 50)

# f: A → B (3 states → 2 states)
f = Simulation(source_id=1, target_id=2, state_map=[0, 1, 0])
print(f"f: Machine 1 → Machine 2")
print(f"   state_map: {f.state_map}")

# g: B → C (2 states → 2 states)
g = Simulation(source_id=2, target_id=3, state_map=[1, 0])
print(f"g: Machine 2 → Machine 3")
print(f"   state_map: {g.state_map}")

# h = f ; g: A → C
h = f.compose(g)
if h:
    print(f"h = f ; g: Machine 1 → Machine 3")
    print(f"   state_map: {h.state_map}")
    print(f"   Verification: state 0 → f[0]={f.state_map[0]} → g[{f.state_map[0]}]={h.state_map[0]}")
````

## 7. Temporal Logic (LTL/CTL)

Temporal logic specifies properties over execution paths.

### LTL (Linear Temporal Logic)

| Operator | Notation | Meaning |
|----------|:--------:|----------|
| Next | X φ | φ holds in the next state |
| Eventually | F φ | φ holds at some future state |
| Globally | G φ | φ holds at all future states |
| Until | φ U ψ | φ holds until ψ becomes true |

### CTL (Computation Tree Logic)

| Operator | Meaning |
|:--------:|----------|
| EX φ | There exists a next state where φ |
| AX φ | In all next states, φ |
| EF φ | There exists a path to φ (reachability) |
| AF φ | All paths eventually reach φ (inevitability) |
| EG φ | Some path maintains φ |
| AG φ | All paths maintain φ (invariant) |

### Property Classification

- **Safety**: "Nothing bad happens" → G ¬bad
- **Liveness**: "Something good eventually happens" → F good

````python
from enum import Enum, auto

class PropertyClass(Enum):
    SAFETY = auto()    # Bad things never happen
    LIVENESS = auto()  # Good things eventually happen
    MIXED = auto()     # Combination


@dataclass
class AtomicProp:
    """An atomic proposition about a state."""
    name: str
    id: int


@dataclass
class TemporalProperty:
    """A temporal property with classification."""
    name: str
    description: str
    classification: PropertyClass
    formula: str  # Human-readable formula


class StandardProperties:
    """Standard temporal properties for state machines."""
    
    @staticmethod
    def deadlock_freedom() -> TemporalProperty:
        return TemporalProperty(
            name="deadlock_freedom",
            description="Non-terminal states can always progress",
            classification=PropertyClass.SAFETY,
            formula="AG (¬terminal → EX true)"
        )
    
    @staticmethod
    def termination() -> TemporalProperty:
        return TemporalProperty(
            name="termination",
            description="All paths eventually reach a terminal state",
            classification=PropertyClass.LIVENESS,
            formula="AF terminal"
        )
    
    @staticmethod
    def reachability(target: str) -> TemporalProperty:
        return TemporalProperty(
            name=f"reachability_{target}",
            description=f"State '{target}' is reachable",
            classification=PropertyClass.LIVENESS,
            formula=f"EF {target}"
        )
    
    @staticmethod
    def invariant(prop: str) -> TemporalProperty:
        return TemporalProperty(
            name=f"invariant_{prop}",
            description=f"Property '{prop}' always holds",
            classification=PropertyClass.SAFETY,
            formula=f"AG {prop}"
        )


# Display standard properties
print("Standard Temporal Properties")
print("=" * 50)

props = [
    StandardProperties.deadlock_freedom(),
    StandardProperties.termination(),
    StandardProperties.reachability("goal"),
    StandardProperties.invariant("safe"),
]

for p in props:
    class_icon = {"SAFETY": "🛡️", "LIVENESS": "⚡", "MIXED": "🔀"}[p.classification.name]
    print(f"\n{class_icon} {p.name}")
    print(f"   {p.description}")
    print(f"   Formula: {p.formula}")
    print(f"   Class: {p.classification.name}")
````

## 8. Equivalence Relations

Different notions of "sameness" for state machines:

| Relation | Strength | Definition |
|----------|:--------:|------------|
| **Bisimulation** | Strongest | Mutual simulation in both directions |
| **Simulation** | Medium | One-directional behavior preservation |
| **Trace Equivalence** | Weakest | Same observable sequences |

### Bisimulation

A relation R ⊆ S₁ × S₂ is a **bisimulation** if:
1. R is a simulation from M₁ to M₂
2. R⁻¹ is a simulation from M₂ to M₁

Bisimilar machines are **observationally indistinguishable**.

````python
@dataclass
class StateRelation:
    """A binary relation on states."""
    pairs: Set[Tuple[int, int]]
    
    def contains(self, s1: int, s2: int) -> bool:
        return (s1, s2) in self.pairs
    
    def is_reflexive(self, states: range) -> bool:
        return all((s, s) in self.pairs for s in states)
    
    def is_symmetric(self) -> bool:
        return all((b, a) in self.pairs for (a, b) in self.pairs)
    
    def is_transitive(self) -> bool:
        for (a, b) in self.pairs:
            for (c, d) in self.pairs:
                if b == c and (a, d) not in self.pairs:
                    return False
        return True
    
    def is_equivalence(self, states: range) -> bool:
        return self.is_reflexive(states) and self.is_symmetric() and self.is_transitive()


@dataclass
class Bisimulation:
    """A bisimulation relation between state machines."""
    relation: StateRelation
    machine1_id: int
    machine2_id: int
    verified: bool = False
    
    @classmethod
    def identity(cls, machine_id: int, state_count: int) -> 'Bisimulation':
        """Identity bisimulation (machine is bisimilar to itself)."""
        pairs = {(s, s) for s in range(state_count)}
        return cls(
            relation=StateRelation(pairs),
            machine1_id=machine_id,
            machine2_id=machine_id,
            verified=True
        )


@dataclass
class BisimulationQuotient:
    """Quotient of a machine by bisimulation equivalence."""
    equivalence_classes: List[Set[int]]
    
    @property
    def state_count(self) -> int:
        """Number of equivalence classes = states in minimal machine."""
        return len(self.equivalence_classes)
    
    def representative(self, class_idx: int) -> int:
        """Get a representative state from an equivalence class."""
        return min(self.equivalence_classes[class_idx])


# Example: Bisimulation quotient
print("Bisimulation Quotient Example")
print("=" * 50)

# Original machine has 6 states, but some are bisimilar
quotient = BisimulationQuotient(
    equivalence_classes=[
        {0, 3},      # States 0 and 3 are bisimilar
        {1, 4},      # States 1 and 4 are bisimilar
        {2, 5},      # States 2 and 5 are bisimilar
    ]
)

print(f"Original states: 6")
print(f"Minimal states (after quotient): {quotient.state_count}")
print(f"\nEquivalence classes:")
for i, cls in enumerate(quotient.equivalence_classes):
    rep = quotient.representative(i)
    print(f"  Class {i}: {cls} (representative: {rep})")
````

## 9. Formal Proofs (Curry-Howard)

The **Curry-Howard correspondence** maps:

| Logic | Type Theory |
|-------|-------------|
| Propositions | Types |
| Proofs | Values |
| Implication (P → Q) | Function (P → Q) |
| Conjunction (P ∧ Q) | Product (P × Q) |
| Disjunction (P ∨ Q) | Sum (P + Q) |
| Universal (∀x.P) | Dependent function |
| Existential (∃x.P) | Dependent pair |

**Key insight**: If you can construct a value of type `Proof<P>`, then P is proven.

````python
from typing import Generic, TypeVar

P = TypeVar('P')  # Proposition type
Q = TypeVar('Q')  # Another proposition

class Proof(Generic[P]):
    """
    A proof witness for proposition P.
    
    The existence of a Proof[P] value witnesses that P is true.
    """
    
    def __init__(self, justification: str = "axiom"):
        self._justification = justification
    
    @property
    def justification(self) -> str:
        return self._justification


class And(Generic[P, Q]):
    """Logical AND: P ∧ Q"""
    
    @staticmethod
    def intro(proof_p: Proof[P], proof_q: Proof[Q]) -> 'Proof[And[P, Q]]':
        """Construct conjunction from proofs of both parts."""
        return Proof(f"And.intro({proof_p.justification}, {proof_q.justification})")
    
    @staticmethod
    def elim_left(proof: 'Proof[And[P, Q]]') -> Proof[P]:
        """Extract left conjunct."""
        return Proof(f"And.elim_left({proof.justification})")
    
    @staticmethod
    def elim_right(proof: 'Proof[And[P, Q]]') -> Proof[Q]:
        """Extract right conjunct."""
        return Proof(f"And.elim_right({proof.justification})")


class Implies(Generic[P, Q]):
    """Logical implication: P → Q"""
    
    @staticmethod
    def modus_ponens(impl: 'Proof[Implies[P, Q]]', proof_p: Proof[P]) -> Proof[Q]:
        """From P → Q and P, derive Q."""
        return Proof(f"modus_ponens({impl.justification}, {proof_p.justification})")


# State machine propositions
class Reachable: pass  # "State S is reachable"
class Terminal: pass   # "State S is terminal"
class Deterministic: pass  # "Machine M is deterministic"
class DeadlockFree: pass   # "Machine M is deadlock-free"


# Example proof
print("Curry-Howard Proof Construction")
print("=" * 50)

# Axioms
proof_reachable: Proof[Reachable] = Proof("reachability_analysis")
proof_terminal: Proof[Terminal] = Proof("terminal_check")

print(f"\nProof of Reachable: {proof_reachable.justification}")
print(f"Proof of Terminal: {proof_terminal.justification}")

# Conjunction
proof_both: Proof[And[Reachable, Terminal]] = And.intro(proof_reachable, proof_terminal)
print(f"\nProof of (Reachable ∧ Terminal): {proof_both.justification}")

# Extraction
proof_reachable_again: Proof[Reachable] = And.elim_left(proof_both)
print(f"\nExtracted Reachable: {proof_reachable_again.justification}")
````

````python
@dataclass
class TheoremSummary:
    """Summary of a formal theorem."""
    id: str
    name: str
    statement: str
    category: str
    prerequisites: List[str]


class TheoremRegistry:
    """Registry of all theorems in the Universal Theory of State."""
    
    THEOREMS = [
        # Preservation Theorems
        TheoremSummary(
            id="T1", name="Bisimulation Preserves CTL",
            statement="Bisimilar machines satisfy the same CTL formulas",
            category="preservation", prerequisites=[]
        ),
        TheoremSummary(
            id="T2", name="Parallel Preserves Safety",
            statement="Product of safe machines is safe",
            category="preservation", prerequisites=["T1"]
        ),
        TheoremSummary(
            id="T3", name="Sequential Preserves Termination",
            statement="Composition of terminating machines terminates",
            category="preservation", prerequisites=[]
        ),
        TheoremSummary(
            id="T4", name="Product Preserves Deadlock Freedom",
            statement="Product of deadlock-free machines is deadlock-free",
            category="preservation", prerequisites=["T2"]
        ),
        TheoremSummary(
            id="T5", name="Refinement Preserves Safety",
            statement="Refinement preserves safety properties",
            category="preservation", prerequisites=["T1"]
        ),
        
        # Soundness Theorems
        TheoremSummary(
            id="S1", name="Typestate Implies L3",
            statement="Typestate pattern guarantees single active state",
            category="soundness", prerequisites=[]
        ),
        TheoremSummary(
            id="S2", name="Method Absence Implies L4",
            statement="Missing methods prevent invalid transitions",
            category="soundness", prerequisites=["S1"]
        ),
        TheoremSummary(
            id="S3", name="Const Generics Imply L11",
            statement="Compile-time bounds prevent structural modification",
            category="soundness", prerequisites=[]
        ),
        
        # Completeness Theorems
        TheoremSummary(
            id="C1", name="Minimal Realization Exists",
            statement="Every finite state machine has a unique minimal realization",
            category="completeness", prerequisites=["T1"]
        ),
        TheoremSummary(
            id="C2", name="Bisimulation Decidable",
            statement="Bisimulation equivalence is decidable in O(n log n) for n states",
            category="completeness", prerequisites=[]
        ),
    ]
    
    @classmethod
    def by_category(cls, category: str) -> List[TheoremSummary]:
        return [t for t in cls.THEOREMS if t.category == category]
    
    @classmethod
    def fundamental(cls) -> List[TheoremSummary]:
        """Theorems with no prerequisites."""
        return [t for t in cls.THEOREMS if not t.prerequisites]
    
    @classmethod
    def dependents(cls, theorem_id: str) -> List[TheoremSummary]:
        """Theorems that depend on the given theorem."""
        return [t for t in cls.THEOREMS if theorem_id in t.prerequisites]


# Display theorem registry
print("Universal Theory of State — Theorem Registry")
print("=" * 60)

for category in ["preservation", "soundness", "completeness"]:
    theorems = TheoremRegistry.by_category(category)
    print(f"\n{category.upper()} THEOREMS")
    print("-" * 40)
    for t in theorems:
        prereq = f" (requires: {', '.join(t.prerequisites)})" if t.prerequisites else " (fundamental)"
        print(f"  {t.id}: {t.name}{prereq}")
        print(f"      {t.statement}")

print(f"\n\nTotal: {len(TheoremRegistry.THEOREMS)} theorems")
print(f"Fundamental: {len(TheoremRegistry.fundamental())}")
````
