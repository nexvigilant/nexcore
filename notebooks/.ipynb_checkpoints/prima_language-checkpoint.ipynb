# Prima (πρίμα) — Primitive-First Programming Language

**Copyright © 2026 NexVigilant LLC. All Rights Reserved.**  
**Intellectual Property of Matthew Alexander Campion, PharmD**

---

A language where **every construct grounds to the 15 Lex Primitiva**. Types are compositions of primitives. Functions are primitive transformations. The runtime tracks grounding and computes transfer confidence.

## Root Constants

All computation ultimately grounds to two mathematical constants:

| Constant | Meaning | Grounding |
|----------|---------|----------|
| **0** | Absence | Mathematical zero |
| **1** | Existence | Mathematical one |

These form the bedrock upon which all higher abstractions are built.

## The 15 Lex Primitiva

```
σ μ ς ρ ∅ ∂ ν ∃ π → κ N λ ∝ Σ
```

| Symbol | Name | Domain | Description |
|--------|------|--------|-------------|
| **σ** | Sequence | Order | Ordered collections, iteration |
| **μ** | Mapping | Transformation | Key-value pairs, functions |
| **ς** | State | Container | Mutable data, variables |
| **ρ** | Recursion | Self-reference | Self-referential structures |
| **∅** | Void | Absence | Null, empty, nothing |
| **∂** | Boundary | Limits | Constraints, bounds |
| **ν** | Frequency | Rate | Counts per time (Greek nu) |
| **∃** | Existence | Creation | Instantiation, allocation |
| **π** | Persistence | Durability | Storage, memory |
| **→** | Causality | Cause-effect | Triggers, dependencies |
| **κ** | Comparison | Predicates | Equality, ordering |
| **N** | Quantity | Numbers | Numeric values |
| **λ** | Location | Address | Pointers, references |
| **∝** | Irreversibility | Consumption | Destructive operations |
| **Σ** | Sum | Variants | Enums, union types |

## Tier System

Types are classified by their **primitive composition complexity**:

| Tier | Primitive Count | Transfer Confidence | Examples |
|------|-----------------|---------------------|----------|
| **T1** | 1 | 1.0 | `N`, `σ`, `μ` |
| **T2-P** | 2-3 | 0.9 | `σ[N]`, `μ[N→N]` |
| **T2-C** | 4-5 | 0.7 | Complex data structures |
| **T3** | 6+ | 0.4 | Domain-specific types |

**Transfer confidence** indicates how reliably concepts map across domains. T1 primitives transfer perfectly (1.0), while T3 domain types have limited transferability (0.4).

## File Extension

Prima source files use the `.σ` extension (sigma), derived from the Sequence primitive:

```
hello.σ
fibonacci.σ
transform.σ
```

ASCII alternative: `.prima`

## Syntax Overview

### Variable Binding (ς)
```prima
let x = 42
let name = "Prima"
```

### Functions (→)
```prima
fn double(x: N) → N {
    x * 2
}

fn add(x: N, y: N) → N {
    x + y
}
```

### Sequences (σ)
```prima
let data = σ[1, 2, 3, 4, 5]
let doubled = data.map(double)
```

### Mappings (μ)
```prima
let pairs = μ("a" → 1, "b" → 2)
```

### Comparison Operators (κ)
```prima
if x κ> 0 { x } else { -x }  // κ> is greater-than
if x κ= y { "equal" }        // κ= is equality
if x κ< 10 { "small" }       // κ< is less-than
```

### Pattern Matching (Σ)
```prima
match result {
    0 → "zero",
    1 → "one",
    _ → "other",
}
```

## Example: Hello World

**File: `hello.σ`**

```prima
// Hello World in Prima
// File: hello.σ
//
// This program demonstrates basic Prima syntax.
// All values ground to the root constants {0, 1}.

println("Hello, Prima!")

// The string "Hello, Prima!" has composition: σ + N
// Tier: T2-P (2 primitives)
```

## Example: Fibonacci

**File: `fibonacci.σ`**

```prima
// Fibonacci in Prima
// File: fibonacci.σ
//
// Demonstrates recursion (ρ) grounded to primitives.

fn fib(n: N) → N {
    if n κ< 2 {
        n
    } else {
        fib(n - 1) + fib(n - 2)
    }
}

let result = fib(10)
println("fib(10) =", result)
println("Tier:", result.tier())
println("Composition:", result.composition())

// fib(10) = 55
// Tier: T1 (single primitive: N)
// Composition: [N]
```

## Example: Data Transformation Pipeline

**File: `transform.σ`**

```prima
// Data Transformation Pipeline in Prima
// File: transform.σ
//
// Demonstrates σ (Sequence) and μ (Mapping) primitives.

fn double(x: N) → N {
    x * 2
}

fn add_one(x: N) → N {
    x + 1
}

// Create a sequence (grounded to σ)
let data = σ[1, 2, 3, 4, 5]

// Apply transformations
let doubled = data.map(double)
let incremented = doubled.map(add_one)

println("Original:", data)
println("Doubled:", doubled)
println("Final:", incremented)

// Check grounding
println("Data tier:", data.tier())
println("Data composition:", data.composition())

// Output:
// Original: σ[1, 2, 3, 4, 5]
// Doubled: σ[2, 4, 6, 8, 10]
// Final: σ[3, 5, 7, 9, 11]
// Data tier: T2-P
// Data composition: [σ + N]
```

## Runtime Composition Tracking

Every Prima value carries its **primitive composition**, enabling:

1. **Runtime tier classification** — Know the complexity of any value
2. **Transfer confidence computation** — Predict cross-domain applicability
3. **Grounding verification** — Prove all values trace to {0, 1}

### Methods Available on Values

| Method | Returns | Description |
|--------|---------|-------------|
| `.tier()` | Tier | T1, T2-P, T2-C, or T3 |
| `.composition()` | [Primitives] | List of constituent primitives |
| `.transfer()` | Float | Transfer confidence (0.0 - 1.0) |

## CLI Usage

```bash
# Start interactive REPL
prima repl

# Run a Prima source file
prima run examples/hello.σ

# Parse and show AST
prima parse examples/fibonacci.σ

# Show tokens
prima tokenize examples/transform.σ

# Show grounding trace
prima trace examples/hello.σ

# Evaluate an expression
prima eval "1 + 2 * 3"
```

## Mathematical Foundation

### Derivation Complexity

The 15 primitives are chosen for **O(1) expressiveness**, not theoretical minimality.

| Set | Size | Expressiveness |
|-----|------|----------------|
| Lambda calculus | 3 | O(n) Church encoding |
| SKI combinators | 3 | O(n²) composition |
| Minimal Lex | 9 | O(n) derivation for 6 concepts |
| **Full Lex** | **15** | **O(1) direct for all** |

### Minimal Set (9 primitives)

```
N, →, σ, μ, ς, ρ, ∅, κ, ∃
```

The 6 "redundant" primitives (Σ, ν, π, λ, ∂, ∝) can be derived but require O(n) composition steps:

| Derived | From | Steps |
|---------|------|-------|
| Σ (Sum) | ∅ + κ | 2 |
| ν (Frequency) | σ + N | 2 |
| π (Persistence) | ς + σ | 2 |
| λ (Location) | N + ς | 2 |
| ∂ (Boundary) | κ + N | 2 |
| ∝ (Irreversibility) | → + ∅ | 2 |

**Conclusion:** 15 optimizes **expressiveness** over theoretical **minimality**.

## Symbol Design Rationale

### Why ν (nu) for Frequency?

The original symbol `f` conflicted with common identifiers (e.g., `fn f(x)`). 

**ν (Greek nu)** is the physics-standard symbol for frequency:
- Wave frequency: ν = c/λ
- Photon energy: E = hν
- No ASCII identifier conflicts

### Unicode Primitive Entry

| Symbol | Unicode | Keyboard Hint |
|--------|---------|---------------|
| σ | U+03C3 | Option+s (Mac) |
| μ | U+03BC | Option+m |

## Crate Structure

```
~/.nexcore/crates/nexcore-prima/
├── Cargo.toml
├── src/
│   ├── lib.rs          # Public API: eval(), parse(), tokenize()
│   ├── token.rs        # Token definitions with primitive symbols
│   ├── lexer.rs        # Tokenizer: μ[σ[char] → σ[Token]]
│   ├── ast.rs          # Abstract Syntax Tree types
│   ├── parser.rs       # Recursive descent parser
│   ├── types.rs        # Type system with compositions
│   ├── value.rs        # Runtime values with tracking
│   ├── interpret.rs    # Tree-walking interpreter
│   ├── builtins.rs     # Built-in functions
│   ├── error.rs        # Error types (grounded in ∂)
│   ├── repl.rs         # Interactive REPL
│   └── bin/
│       └── prima.rs    # CLI binary
└── examples/
    ├── hello.σ
    ├── fibonacci.σ
    └── transform.σ
```

## Summary

**Prima** is a primitive-first programming language where:

1. **Every value** carries its primitive composition
2. **Every type** is classified by tier (T1/T2-P/T2-C/T3)
3. **Every computation** grounds to {0, 1}
4. **Transfer confidence** predicts cross-domain applicability

The **15 Lex Primitiva** provide O(1) expressiveness for all common programming concepts:

```
σ μ ς ρ ∅ ∂ ν ∃ π → κ N λ ∝ Σ
```

**File extension:** `.σ` (sigma)

**Root constants:** `0` (absence), `1` (existence)
